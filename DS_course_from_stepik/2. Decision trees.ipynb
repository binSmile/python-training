```python
EY = -4/10 * np.log2(4/10) - 6/10 * np.log2(6/10)
print(EY)
```

```python
ig1 = EY - 0.9 * E_fur_true - 0.1 * E_fur_false
ig2 = EY - 0.5 * E_gav_true - 0.5 * E_gav_false
ig3 = EY - (0)
print('Шерстист: %.2f' % ig1)
print('Гав: %.2f' % ig2)
print('Лаз: %.2f' % ig3)
```
